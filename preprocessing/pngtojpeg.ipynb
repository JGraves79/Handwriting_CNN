```python
from PIL import Image

# Load the image
img_path = '/content/mnist_paragraph3.png'
img = Image.open(img_path)

# Convert to JPEG and save
jpeg_path = 'mnsit_paragraph3.jpeg'
img = img.convert('RGB')
img.save(jpeg_path, 'JPEG')

print(f"Image successfully converted and saved as {jpeg_path}")
```
