```python
import os
from PIL import Image

def convert_png_to_jpeg(folder_path, target_folder):
    # Ensure the folder exists and is writable
    if not os.path.exists(folder_path):
        print(f"Error: The directory {folder_path} does not exist.")
        return
    if not os.access(folder_path, os.W_OK):
        print(f"Error: The directory {folder_path} is not writable.")
        return

    # Iterate through all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file is a PNG and ends with "r01"
        if filename.lower().endswith('r01.png'):
            try:
                # Construct the full path to the PNG file
                png_path = os.path.join(folder_path, filename)
                # Open the PNG file
                img = Image.open(png_path)
                # Convert to JPEG
                img = img.convert('RGB')
                # Construct the JPEG file path
                jpeg_filename = os.path.splitext(filename)[0] + '.jpeg'
                jpeg_path = os.path.join(target_folder, jpeg_filename)
                # Save the image in JPEG format
                img.save(jpeg_path, 'JPEG')
                print(f"Image successfully converted and saved as {jpeg_path}")
            except PermissionError as e:
                print(f"PermissionError: {e}")
            except Exception as e:
                print(f"An error occurred: {e}")

# Usage
folder_path = 'c:\\Users\\X579430\\OneDrive - Nissan Motor Corporation\\Documents\\Bootcamp\\Handwriting_CNN\\Source_Data'
target_folder = 'c:\\Users\\X579430\\OneDrive - Nissan Motor Corporation\\Documents\\Bootcamp\\Handwriting_CNN\\Source_jpeg'
convert_png_to_jpeg(folder_path, target_folder)
```
